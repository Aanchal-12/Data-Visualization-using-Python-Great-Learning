```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
%matplotlib inline
```

```
auto = pd.read_csv('Automobile.csv')
```

```
auto.head()
```

```
sns.distplot(auto['normalized_losses'])
plt.show()
```

```
sns.distplot(auto['city_mpg'])
plt.show()
```

```
sns.distplot(auto['city_mpg'], kde=False, rug= True)
plt.show()
```

```
sns.jointplot(x=auto['engine_size'], y=auto['horsepower'])
plt.show()
```

```
sns.jointplot(x=auto['engine_size'], y=auto['horsepower'], kind="hex", marginal_ticks=True)
plt.show()
```

```
sns.jointplot(x=auto['engine_size'], y=auto['horsepower'], kind="resid")
plt.show()
```

```
sns.jointplot(x=auto['engine_size'], y=auto['horsepower'], kind="reg")
plt.show()
```

```
sns.jointplot(x=auto['engine_size'], y=auto['horsepower'], kind="kde")
```

```
sns.pairplot(auto[['normalized_losses', 'engine_size', 'horsepower']])
```

```
sns.stripplot(x= auto['fuel_type'], y =auto['horsepower'])
```

```
sns.stripplot(x=auto['fuel_type'], y=auto['horsepower'], jitter=True)
```

```
sns.swarmplot(x=auto['fuel_type'], y=auto['horsepower'])
```

```
sns.boxplot(x=auto['number_of_doors'], y=auto['horsepower'])
```

```
sns.boxplot(x=auto['number_of_doors'], y=auto['horsepower'], hue=auto['fuel_type'])
```

```
sns.barplot(x=auto['body_style'], y=auto['horsepower'])
```

```
sns.barplot(x=auto['body_style'], y=auto['horsepower'], hue=auto['engine_location'])
```

```
sns.countplot(auto['body_style'])
```

```
sns.countplot(auto['body_style'], hue=auto['engine_location'])
```

```
sns.pointplot(x= auto['fuel_system'], y= auto['horsepower'])
```

```
sns.pointplot(x= auto['fuel_system'], y= auto['horsepower'], hue= auto['number_of_doors'])
```

```
sns.catplot(x= 'fuel_type', y= 'horsepower', hue= 'number_of_doors', col = 'engine_location', data= auto, kind = 'swarm')
```

```
sns.catplot(x= 'fuel_type', y= 'horsepower', hue= 'number_of_doors', col = 'engine_location', data= auto, kind = 'violin')
```

```
sns.catplot(x= 'fuel_type', y= 'horsepower', hue= 'number_of_doors', col = 'engine_location', data= auto, kind = 'strip')
```

```
sns.lmplot(x='horsepower', y='peak_rpm', data=auto)
```

```
sns.lmplot(x='horsepower', y='peak_rpm', data=auto, hue='fuel_type')
```
